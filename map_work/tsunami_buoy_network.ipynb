```python
# Working Code for Plot Generation
# Define the Indian Ocean Region region
region = [60, 100, 0, 30]

# Tsunami Buoy locations 
stb_buoys = {
    "name": ["STB-01", "STB-02"],
    "latitude": [6.25, 20.80],
    "longitude": [88.88, 64.34],
    }

itb_buoys = {
    "name": ["ITB-05", "ITB-06", "ITB-09", "ITB-12"], 
    "latitude": [10.17, 13.71, 16.94, 20.18],
    "longitude": [88.49, 89.56, 89.99, 67.65],
}

# Create the figure
fig = pygmt.Figure()

# Draw the base map
fig.coast(
    region=region,
    projection="M15c",  # Mercator projection
    land="lightgray",
    water="lightblue",
    shorelines="1/0.5p,black",
    frame=["af", "+tIndian Tsunami Buoy Network"],
    borders="1/0.5p,black",
    resolution="h"
)

# Plot STB Tsunami Buoys Location

fig.plot(
    x= stb_buoys["longitude"],
    y= stb_buoys["latitude"], 
    style="t0.4c",  # Triangle symbol
    fill="red",
    pen="black",
    label="INCOIS Network"
    )

# Plot STB Tsunami Buoys Labels

fig.text(
    x= [i+1 for i in stb_buoys["longitude"]],
    y= [i+1 for i in stb_buoys["latitude"]],
    text=stb_buoys["name"],
    font="10p,Helvetica-Bold,black"
    )


# Plot ITB Tsunami Buoys Locations

fig.plot(
    x=itb_buoys["longitude"],
    y=itb_buoys["latitude"],
    style="c0.4c",  # Triangle symbol
    fill="green",
    pen="black",
    label="NIOT Network"
    )

# Plot ITB Tsunami Buoys Labels

fig.text(
    x=[i+1 for i in itb_buoys["longitude"]],
    y=[i+1 for i in itb_buoys["latitude"]],
    text=itb_buoys["name"],
    font="10p,Helvetica-Bold,black",
    )

fig.legend(position="JTL+jTL+o0.5c", box=True)

fig.savefig("indian_tsunami_buoy_network2.jpeg", dpi = 500)
fig.show()
```

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# Define the Indian Ocean Region extent
extent = [60, 100, 0, 30]

# Tsunami Buoy locations
stb_buoys = {
    "name": ["STB-01", "STB-02"],
    "latitude": [6.25, 20.80],
    "longitude": [88.88, 64.34],
}

itb_buoys = {
    "name": ["ITB-05", "ITB-06", "ITB-09", "ITB-12"],
    "latitude": [10.17, 13.71, 16.94, 20.18],
    "longitude": [88.49, 89.56, 89.99, 67.65],
}

# Create figure and Mercator projection
fig = plt.figure(figsize=(10, 8))
ax = plt.axes(projection=ccrs.Mercator())
ax.set_extent(extent, crs=ccrs.PlateCarree())
ax.set_title("Indian Tsunami Buoy Network", fontsize=14, weight='bold')

# Add map features
ax.add_feature(cfeature.LAND, facecolor='lightgray')
ax.add_feature(cfeature.OCEAN, facecolor='lightblue')
ax.add_feature(cfeature.COASTLINE, linewidth=0.5)
ax.add_feature(cfeature.BORDERS, linewidth=0.5)
ax.gridlines(draw_labels=True, linewidth=0.5, color='gray', alpha=0.5, linestyle='--')

# Plot STB buoys
ax.plot(
    stb_buoys["longitude"], stb_buoys["latitude"],
    marker="^", color="red", markeredgecolor="black",
    linestyle="", transform=ccrs.PlateCarree(), label='STB Buoys'
)

# Plot STB labels
for lon, lat, name in zip(stb_buoys["longitude"], stb_buoys["latitude"], stb_buoys["name"]):
    ax.text(lon + 1, lat + 1, name, fontsize=10, fontweight='bold',
            transform=ccrs.PlateCarree(), color='black')

# Plot ITB buoys
ax.plot(
    itb_buoys["longitude"], itb_buoys["latitude"],
    marker="o", color="green", markeredgecolor="black",
    linestyle="", transform=ccrs.PlateCarree(), label='ITB Buoys'
)

# Plot ITB labels
for lon, lat, name in zip(itb_buoys["longitude"], itb_buoys["latitude"], itb_buoys["name"]):
    ax.text(lon + 1, lat + 1, name, fontsize=10, fontweight='bold',
            transform=ccrs.PlateCarree(), color='black')

# Show legend and plot
plt.legend(loc= "lower left")
plt.savefig("tsunami_buoy_network_1.png")
plt.show()
```
