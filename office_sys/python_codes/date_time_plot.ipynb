```python
import pygmt
```

```python
data = pygmt.datasets.load_sample_data(name="japan_quakes")

region = [
    data.longitude.min() - 1,
    data.longitude.max() + 1,
    data.latitude.min() - 1,
    data.latitude.max() + 1,
]

print(region)
print(data.head())
```

```python
fig = pygmt.Figure()
fig.basemap(region=region, projection="M15c", frame=True)
fig.coast(land="black", water="skyblue")
pygmt.makecpt(cmap="viridis", series=[data.depth_km.min(), data.depth_km.max()])
fig.plot(
    x=data.longitude, 
    y=data.latitude, 
    size = 0.015 * (2**data.magnitude),
    cmap = True,
    style="cc", 
    fill=data.depth_km,#"white", 
    pen="black"
)

fig.colorbar(frame="xaf+lDepth (km)")
fig.show()
```

```python
import datetime

import numpy as np
import pandas as pd
import pygmt
import xarray as xr
```

```python
#x = pd.date_range("2018-03-01", periods=7, freq="BQS")
x = pd.date_range(
    start ='16-09-2023',  
    end ='17-10-2023',
    freq='D',
)
y = np.ones(len(x))

fig = pygmt.Figure()
fig.plot(
    projection="X10c/10c",
    region=[datetime.datetime(2023, 9, 16), datetime.datetime(2023, 9, 17), 0, 5],
    frame=["WSen", "ag"],
    x=x,
    y=y,
    style="i0.4c",
    pen="1p",
    fill="purple",
)
fig.show()
```
