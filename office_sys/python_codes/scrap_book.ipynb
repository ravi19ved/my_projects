```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Seed for reproducibility
np.random.seed(2)

# Data
data = np.random.random((8, 8))

# Heat map
fig, ax = plt.subplots()
ax.imshow(data)
# plt.show()
```

```python
# Labels

xlabs = ["2010", "2011", "2012", "2013",
         "2014", "2015", "2016", "2017"]
ylabs = ["Aerial Bay", "Campbell Bay", "Chennai", "Ennore",
         "Kochi", "Beypore", "Nagapattinam", "Kollam"]

         
# Heat map
fig, ax = plt.subplots()
ax.imshow(data)

# Add the labels
ax.set_xticks(np.arange(len(xlabs)), labels = xlabs)
ax.set_yticks(np.arange(len(ylabs)), labels = ylabs)
# plt.show()
```
