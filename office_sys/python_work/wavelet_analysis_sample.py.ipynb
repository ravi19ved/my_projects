```python
import numpy as np
import matplotlib.pyplot as plt
import pywt
import pandas as pd

# Load tide gauge data
data_raw = pd.read_csv("tide_data.txt", names=['date', 'time', 'value'], delimiter='\s+') # Replace 'tide_data.txt' with your actual data file
data_raw['datetime'] = data_raw['date'] + " " + data_raw['time']
data_raw.drop(['date', 'time'], axis=1, inplace=True)
data_raw['datetime'] = pd.to_datetime(data_raw.datetime, format='%m-%d-%Y %H:%M')
# print(data_raw.head())

datetime_np = data_raw['datetime'].to_numpy()
values_np = data_raw['value'].to_numpy()

data = [datetime_np, values_np]
data = (np.array(data)).T

# Extract time and tide values
time = data[:, 0]
tide = data[:, 1]


# Perform wavelet analysis
wavelet = 'db4'  # Choose the wavelet function (e.g., Daubechies 4)
level = 6  # Choose the number of decomposition levels

coeffs = pywt.wavedec(tide, wavelet, level=level)
cA = coeffs[0]  # Approximation coefficients
cD = coeffs[1:]  # Detail coefficients

# Plot the original tide data and its wavelet coefficients
plt.figure(figsize=(10, 6))
plt.subplot(level+2, 1, 1)
plt.plot(time, tide)
plt.title('Original Tide Data')

for i, c in enumerate(cD):
    plt.subplot(level+2, 1, i+2)
    plt.plot(time, c)
    plt.title('Detail Coefficients: Level {}'.format(i+1))

plt.subplot(level+2, 1, level+2)
plt.plot(time, cA)
plt.title('Approximation Coefficients')

plt.tight_layout()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
import pywt

# Load tide gauge data
data = np.loadtxt('tide_data.txt', dtype = 'str', delimiter = ' ')  # Replace 'tide_data.txt' with your actual data file

# Extract time and tide values
time = data[:, 0]
tide = data[:, 1]

# Define the wavelet function and scales
wavelet = 'morl'  # Choose the wavelet function (e.g., Morlet)
scales = np.arange(1, 101)  # Define the scales for the analysis

# Perform continuous wavelet transform
coeffs, freqs = pywt.cwt(tide, scales, wavelet, dt = 1)  # dt is the time step

# Plot the continuous wavelet transform
plt.figure(figsize=(10, 6))
plt.imshow(np.abs(coeffs), extent=[time[0], time[-1], freqs[-1], freqs[0]], cmap='jet', aspect='auto')
plt.colorbar(label='Wavelet Coefficient')
plt.title('Continuous Wavelet Transform of Tide Data')
plt.xlabel('Time')
plt.ylabel('Scale')
plt.gca().invert_yaxis()
plt.show()
```
