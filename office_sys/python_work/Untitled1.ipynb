```python
import folium

def generate_map(points):
    # Create a map centered on the first point in the list
    map_center = points[0]
    map_object = folium.Map(location=map_center, zoom_start=12)

    # Add markers for each point on the map
    for point in points:
        folium.Marker(point).add_to(map_object)

    # Save the map as an HTML file
    map_object.save('map.html')

# Example usage
points = [
    (51.5074, -0.1278),  # London
    (40.7128, -74.0060),  # New York
    (48.8566, 2.3522),  # Paris
    (35.6895, 139.6917),  # Tokyo
]

generate_map(points)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pywt
```

```python
#Generate some sample tide gauge data

time = np.linspace(0, 10, 1000)
signal = np.sin(2 * np.pi * time) + 0.5 * np.sin(2 * np.pi * 1.5 * time) + 0.8 * np.sin(2 * np.pi * 1.2 * time)

#Performing wavelet analysis
wavelet = 'db4' # Choose the wavelet function (e.g. Doubechies 4)
level = 6 # Choose the number of decomposition levels

coeffs = pywt.wavedec(signal, wavelet, level = level)

cA = coeffs[0]     # Approximation coefficients
cD = coeffs[1:]     # Detail coefficients

print(len(cA))
print(len(cD))
plt.figure(figsize=(10, 6))
plt.plot(time, signal)
plt.show()
```
