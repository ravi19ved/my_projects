```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load tide data from a CSV file into a pandas DataFrame
df = pd.read_csv('Cochin_ENC_2016_hourly.csv')

df.columns =['DateTime', 'Tide']
df['DateTime'] = pd.to_datetime(df['DateTime'], format="%m-%d-%Y %H:%M")
df['DateTime'] = df['DateTime'].sort_index(0)


# Plotting the tide data
plt.figure(figsize=(18, 6))
df.iloc[901:2700].plot(y='Tide')
#plt.plot(df['DateTime'], df['Tide'], color='blue')
plt.xlabel('Date & Time')
plt.ylabel('Tide Level')
plt.title('Tide Data Plot')
plt.grid(True)
plt.show()
```
