```python
new_reg = [70,75,15,25]
grid = pygmt.datasets.load_earth_relief(resolution="30s", region = new_reg, registration="gridline")

fig = pygmt.Figure()
fig.grdimage(grid=grid, frame="a", projection = "M20c", cmap = "grayC")
fig.grdcontour(grid=grid, interval=200, annotation=500)
fig.coast(shorelines="2p", land="lightgray")
fig.colorbar(frame=["a1000", "x+lElevation", "y+lm"])
fig.show()
```
