```python
import pygmt

# Load sample earth relief data
grid = pygmt.datasets.load_earth_relief(resolution="01m", region=[65,75,15,25])
```

```python
fig = pygmt.Figure()
fig.grdcontour(grid=grid, projection = "M10c",limit=[-4000, -250],
interval=250, annotation=1000, frame=["a","+tBathy Contours"])
fig.coast(shorelines="1p", land="lightgray")
fig.show()
```
