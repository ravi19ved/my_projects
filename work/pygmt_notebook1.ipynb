```python
import pygmt
#pygmt.show_versions()
```

```python
fig = pygmt.Figure()

region = [-180, 180, -60, 60]

fig.coast(shorelines = "1/0.5p", region = region, projection = "M25c", frame = True)
fig.basemap(frame=["a30f7.5g15","+tWorld Map"])
fig.show()
```
