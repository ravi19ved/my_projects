```python
import pygmt
import numpy as np

# Dummy data for gliders
data = {
    "glider_id": ["G1", "G2", "G3", "G4", "G5", "G6", "G7", "G8", "G9", "G10"],
    "deployment_lat": [30, 28, 25, 22, 20, 18, 15, 10, 5, 3],
    "deployment_lon": [70, 72, 75, 78, 80, 83, 85, 88, 89, 90],
    "retrieval_lat": [20, 18, 15, 12, 10, 8, 5, 2, 1, 0],
    "retrieval_lon": [70, 72, 75, 78, 80, 83, 85, 88, 89, 90],
}

# Generate random intermediate positions
def generate_random_positions(start_lat, start_lon, end_lat, end_lon, n=10):
    lats = np.linspace(start_lat, end_lat, num=n)
    lons = np.linspace(start_lon, end_lon, num=n)
    return lats, lons

# Create a PyGMT figure
fig = pygmt.Figure()

# Define the map region and projection
region = [60, 98, -10, 25]  # [west, east, south, north]
projection = "M6i"  # Mercator projection

# Plot the base map
fig.basemap(region=region, projection=projection, frame=["af", "+t'Glider Transects in the Northern Indian Ocean'"])

# Add coastlines and political boundaries
fig.coast(shorelines="1/0.5p,black", resolution="1/0.5p,black", water="lightblue")

# Plot the glider transects
for i, glider_id in enumerate(data["glider_id"]):
    start_lat = data["deployment_lat"][i]
    start_lon = data["deployment_lon"][i]
    end_lat = data["retrieval_lat"][i]
    end_lon = data["retrieval_lon"][i]
    
    # Generate dummy positions
    lats, lons = generate_random_positions(start_lat, start_lon, end_lat, end_lon)
    
    # Plot the transect
    fig.plot(x=lons, y=lats, pen="1p,red", label=f"Glider {glider_id}")
    
    # Plot deployment and retrieval points
    fig.plot(x=start_lon, y=start_lat, style="c0.4c", color="green", pen="0.5p,black", label=f"Deployment {glider_id}")
    fig.plot(x=end_lon, y=end_lat, style="c0.4c", color="red", pen="0.5p,black", label=f"Retrieval {glider_id}")
    
    # Add text labels
    fig.text(x=[start_lon, end_lon], y=[start_lat, end_lat], text=[f"D-{glider_id}", f"R-{glider_id}"], font="8p,black", justify="LM")

# Add a legend
#fig.legend()

# Show the plot
fig.show()
```
