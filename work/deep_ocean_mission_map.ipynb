```python
# Load Argo Floats data from a CSV file (replace 'argo_floats.csv' with the actual file path)
argo_floats = pd.read_csv("argo_floats.csv")

# Load Wave Drifters data from a CSV file (replace 'wave_drifters.csv' with the actual file path)
wave_drifters = pd.read_csv("wave_drifters.csv")

# List of glider transect CSV file paths (replace with actual file paths)
glider_transect_files = [
    "track_890.csv",
    "track_890dep.csv",
    "track_891.csv",
    "track_891dep.csv",
    "track_1096.csv",
    "track_1127.csv",
    "track_1128.csv",
    "track_1130.csv",
    "track_1131.csv"  
]

# Initialize the map
fig = pygmt.Figure()
fig.basemap(region=[46, 110, -20, 28], projection="M8i", frame="af")
fig.coast(land="lightgray", water="skyblue", borders=[1, 2], shorelines="1/thinner")

# Add map title
#fig.text(text="Deep Ocean Mission Deployed Ocean Observation Platforms", position="TC", font="14p,Helvetica-Bold,black")

# Plot Argo Floats
fig.plot(
    x=argo_floats["longitude"],
    y=argo_floats["latitude"],
    style="i0.4c", color="blue", pen="black", label="Argo Floats"
)

# Plot Wave Drifters
fig.plot(
    x=wave_drifters["longitude"],
    y=wave_drifters["latitude"],
    style="c0.4c", color="orange", pen="black", label="Wave Drifters"
)

# Plot Glider Transects
for file_path in glider_transect_files:
    mission_name = file_path.split("_")[-1].replace(".csv", "")  # Extract mission name from file name
    transect_data = pd.read_csv(file_path)
    fig.plot(
        x=transect_data["Longitude"],
        y=transect_data["Latitude"],
        pen="2p,red", label=mission_name
    )



# Add legend
fig.legend(position="JBR+jBR+o0.5c", box=True)

# Save figure
#fig.savefig("full_deployments_map_0.jpeg")

# Show the map
fig.show()
```
