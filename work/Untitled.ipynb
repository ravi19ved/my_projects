```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate dummy data
np.random.seed(42)  # For reproducibility
latitudes = np.linspace(5, 20, 100)  # Latitude range for Bay of Bengal
depths = np.linspace(0, 1000, 100)  # Depth range in meters

# Create a grid of latitude and depth
lat_grid, depth_grid = np.meshgrid(latitudes, depths)

# Generate dummy temperature and salinity data
temperature = 25 - 0.01 * depth_grid + 0.1 * np.sin(0.1 * lat_grid) + np.random.normal(0, 0.5, lat_grid.shape)
salinity = 34 + 0.005 * depth_grid - 0.05 * np.sin(0.1 * lat_grid) + np.random.normal(0, 0.1, lat_grid.shape)

# Plot the meridional TS diagram
plt.figure(figsize=(10, 8))
sns.set_style("whitegrid")

# Scatter plot of Temperature vs Salinity, colored by depth
sc = plt.scatter(salinity.flatten(), temperature.flatten(), c=depth_grid.flatten(), cmap='viridis', s=10)

# Add colorbar
cbar = plt.colorbar(sc)
cbar.set_label('Depth (m)')

# Add labels and title
plt.xlabel('Salinity (PSU)')
plt.ylabel('Temperature (°C)')
plt.title('Meridional TS Diagram in the Bay of Bengal')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate dummy data
np.random.seed(42)  # For reproducibility
latitudes = np.linspace(5, 20, 100)  # Latitude range for Bay of Bengal (5°N to 20°N)
depths = np.linspace(0, 1000, 100)  # Depth range in meters (0 to 1000 m)

# Create a grid of latitude and depth
lat_grid, depth_grid = np.meshgrid(latitudes, depths)

# Generate dummy temperature and salinity data
temperature = 25 - 0.01 * depth_grid + 0.1 * np.sin(0.1 * lat_grid) + np.random.normal(0, 0.5, lat_grid.shape)
salinity = 34 + 0.005 * depth_grid - 0.05 * np.sin(0.1 * lat_grid) + np.random.normal(0, 0.1, lat_grid.shape)

# Create the plot
plt.figure(figsize=(12, 8))

# Plot temperature as a filled contour
plt.subplot(2, 1, 1)  # First subplot for temperature
cf1 = plt.contourf(lat_grid, depth_grid, temperature, levels=20, cmap='plasma')
plt.colorbar(cf1, label='Temperature (°C)')
plt.gca().invert_yaxis()  # Invert y-axis to show depth increasing downward
plt.xlabel('Latitude (°N)')
plt.ylabel('Depth (m)')
plt.title('Temperature vs Latitude and Depth')

# Plot salinity as a filled contour
plt.subplot(2, 1, 2)  # Second subplot for salinity
cf2 = plt.contourf(lat_grid, depth_grid, salinity, levels=20, cmap='viridis')
plt.colorbar(cf2, label='Salinity (PSU)')
plt.gca().invert_yaxis()  # Invert y-axis to show depth increasing downward
plt.xlabel('Latitude (°N)')
plt.ylabel('Depth (m)')
plt.title('Salinity vs Latitude and Depth')

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```
